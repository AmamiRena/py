```python
def get_variety(data):
    sets=[tuple(i) for i in data]
    num_variety=len(set(sets))
    return num_variety

get_variety(data),data[20]
```

```python
def d_euc(x,y):
    d=np.sqrt(np.sum(np.square(x-y)))
    return d
```

```python
def center_update(clusters,data,centers):
    num_centers=np.shape(centers)[0]
    num_features=np.shape(centers)[1]
    container=[list() for x in range(num_centers)]
    for i,cluster in enumerate(clusters):
        container[cluster].append(data[i])
    
    container=np.array(list(map(lambda x:np.array(x),container)))
    new_centers=np.array([])
    
    for i in range(len(container)):
        each_center=np.mean(container[i],axis=0)
        new_centers=np.append(new_centers,each_center)
    
    return new_centers.reshape(num_centers,num_features)
```

```python
def kmeans_cluster(data,init_centers,k,max_step=50,epsilon=.001):
    old_centers=init_centers
    centers_container,cluster_container=[],[]
    centers_container.append(old_centers)
    
    for s in range(max_step):
        cluster=np.array([],dtype=int)
        for ea in data:
            distances=np.array([])
            for ea_c in old_centers:
                temp_distance=d_euc(ea,ea_c)
                distances=np.append(distances,temp_distance)
            lab=np.argmin(distances)
            cluster=np.append(cluster,lab)
        cluster_container.append(cluster)
        
        new_centers=center_update(cluster,data,old_centers)
        
        difference=[]
        for ea_o,ea_n in zip(old_centers,new_centers):
            difference.append(d_euc(ea_o,ea_n))
        
        if (np.array(difference)<epsilon).all():
            return centers_container,cluster_container
        
        centers_container.append(new_centers)
        old_centers=new_centers
    
    return centers_container,cluster_container
```

```python
def randomK(k,data):
    prng=np.random.RandomState(0)
    num_feature=np.shape(data)[1]
    init_centers=prng.randn(k,num_feature)*5
    return init_centers
```
