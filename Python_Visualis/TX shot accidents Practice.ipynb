```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

fontsize=20
plt.rcParams['axes.titlesize']=fontsize
plt.rcParams['legend.frameon']=False
```

```python
from pywaffle import Waffle
import numpy as np

shots={'Fatal':247,'Nonfatal':239,'Missed Shots':152,'Unknown':656-247-239-152}
values=[num for reason,num in shots.items()]
sums=np.sum(values)

plt.figure(FigureClass=Waffle,rows=16,values=values,labels=["%s:%.1f%%\n%s:%d" % (reason, 100*num/sums, 'Nums', num) for reason, num in shots.items()],
           legend={'bbox_to_anchor': (0, -0.2), 'fontsize': 15,'ncol': 4, 'loc': 'lower left', },figsize=(12,8))
plt.title('Shoting accidents in TX',fontsize=25)
```

### Liquid Drop Chart

```python
from pyecharts import options as opts
from pyecharts.charts import Liquid

liquid=Liquid()
liquid.add('Shot Accidents',[.41,.2,.12],label_opts=opts.LabelOpts(is_show=True,position='inside'),is_outline_show=False)
liquid.set_global_opts(title_opts=opts.TitleOpts(title='The black involved in shot accidents in TX',pos_left='30%'))
liquid.render_notebook()
```

```python
from pyecharts.charts import Grid

police={'Fired': 10, 'Suspended': 14, 'Written reprimand': 1}
values = np.round([value for value in police.values()]/np.sum([value for value in police.values()]), 2)
keys=[key for key in police.keys()]
xpos=['14%','45%','76%']
keyxpos=['10%','37%','65%']
grid=Grid()

for xpo,key,value,keyxpo in zip(xpos,keys,values,keyxpos):
    liquid=(Liquid().add(key,[value],center=[xpo,'50%'],is_outline_show=False).set_global_opts(title_opts=opts.TitleOpts(title=key,pos_left=keyxpo)))
    grid.add(liquid,grid_opts=opts.GridOpts())
grid.render_notebook()
```
